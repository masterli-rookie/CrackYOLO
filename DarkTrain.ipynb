```python
import torch
from ultralytics import YOLO
import warnings
import random
import numpy as np
import os

warnings.filterwarnings('ignore')

# 设置随机种子
def set_seed(seed=42):
    """设置所有随机种子以确保可重复性"""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)  # 如果使用多GPU
    
    # 设置cuDNN以获得确定性结果（可能会降低性能）
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    
    # 设置Python哈希种子
    os.environ['PYTHONHASHSEED'] = str(seed)
    
    print(f"随机种子已设置为: {seed}")

# 模型配置文件
model_yaml_path = r'../ultralytics-main/ultralytics/cfg/models/addv11/yolov11n.yaml'
# 数据集配置文件
data_yaml_path = r'../ultralytics-main/your_dataset/data.yaml'

if __name__ == '__main__':
    # 设置随机种子（可以调整这个值）
    set_seed(42)
    
    # 初始化模型
    model = YOLO(model_yaml_path)
    
    # 训练模型
    results = model.train(
        data=data_yaml_path,
        imgsz=640,
        epochs=300,
        batch=32,
        workers=0,
        amp=False,  # 如果出现训练损失为Nan可以关闭amp
        project='runs/V11train',
        name='exp_dark_no_aug',  # 建议修改实验名，方便区分
        deterministic=True,  # 确保确定性行为
        seed=42,  # 传递给训练器的种子
        
        # ====== 以下为关闭光照/色彩增强的设定 ======
        hsv_h=0.0,           # 关闭色调增强
        hsv_s=0.0,           # 关闭饱和度增强
        hsv_v=0.0,           # 关闭亮度增强 (核心！防止暗图被随机提亮)
        auto_augment=None,   # 关闭 RandAugment (防止触发内部的亮度/均衡化操作)
        erasing=0.0,         # 关闭随机擦除 (暗光下目标本就模糊，擦除容易丢失关键特征)
        # ==========================================
    )
```
